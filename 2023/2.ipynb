```python
# # Input: Multi-line string representing games
# games_input = """
# Game 1: 3 blue, 4 red; 1 red, 2 green, 6 blue; 2 green
# Game 2: 6 red, 1 blue, 3 green; 2 blue, 1 red, 2 green
# Game 3: 4 red, 5 blue; 3 green, 2 red, 4 blue
# Game 4: 2 red, 1 blue; 1 green, 2 blue, 3 red
# Game 5: 6 red, 1 blue, 3 green; 2 blue, 1 red, 2 green
# """

def parse_game(game_str):
    parts = game_str.split(":")
    game_id = parts[0].strip()
    actions = parts[1].strip().split(";")
    return game_id, actions

def count_balls(actions):
    ball_count = {'red': 0, 'blue': 0, 'green': 0}
    for action in actions:
        returned_balls = {color: 0 for color in ball_count}
        for info in action.split(","):
            quantity, color = info.strip().split()
            returned_balls[color] += int(quantity)
        for color in ball_count:
            ball_count[color] = max(ball_count[color], returned_balls[color])
    return ball_count

def find_matching_games(games_input):
    matching_games = []
    for game_str in games_input.strip().split("\n"):
        game_id, actions = parse_game(game_str)
        ball_count = count_balls(actions)
        if ball_count['red'] <= 12 and ball_count['green'] <= 13 and ball_count['blue'] <= 14:
            matching_games.append(game_id)
    return matching_games

matching_games = find_matching_games(games_input)
matching_games_indices = [int(game_id.split()[1]) for game_id in matching_games]

sum_of_indices = sum(matching_games_indices)
print("Matching Games:", matching_games)
print("Sum of Indices:", sum_of_indices)
```

```python
# # Input: Multi-line string representing games
# games_input = """
# Game 1: 3 blue, 4 red; 1 red, 2 green, 6 blue; 2 green
# Game 2: 6 red, 1 blue, 3 green; 2 blue, 1 red, 2 green
# Game 3: 4 red, 5 blue; 3 green, 2 red, 4 blue
# Game 4: 2 red, 1 blue; 1 green, 2 blue, 3 red
# Game 5: 6 red, 1 blue, 3 green; 2 blue, 1 red, 2 green
# """

def parse_game(game_str):
    parts = game_str.split(":")
    game_id = parts[0].strip()
    actions = parts[1].strip().split(";")
    return game_id, actions

def count_balls(actions):
    ball_count = {'red': 0, 'blue': 0, 'green': 0}
    for action in actions:
        returned_balls = {color: 0 for color in ball_count}
        for info in action.split(","):
            quantity, color = info.strip().split()
            returned_balls[color] += int(quantity)
        for color in ball_count:
            ball_count[color] = max(ball_count[color], returned_balls[color])
    return ball_count

def calculate_product(game_str):
    _, actions = parse_game(game_str)
    ball_count = count_balls(actions)
    product = ball_count['red'] * ball_count['green'] * ball_count['blue']
    return product

def sum_products(games_input):
    total_sum = 0
    for game_str in games_input.strip().split("\n"):
        total_sum += calculate_product(game_str)
    return total_sum

total_sum = sum_products(games_input)
print("Sum of Products:", total_sum)
```
