```python
# input_d = '''
# #.##..##.
# ..#.##.#.
# ##......#
# ##......#
# ..#.##.#.
# ..##..##.
# #.#.##.#.

# #...##..#
# #....#..#
# ..##..###
# #####.##.
# #####.##.
# ..##..###
# #....#..#
# '''
def to_add_list(input_d):
    rows = input_d.strip().split('\n')
    return [[*r] for r in rows]
        
def is_valid(list_len, idx):
    return idx >= 0 and idx < list_len

def get_col_list(block_2d, c_idx):
    return [r[c_idx] for r in block_2d]

def get_row_list(block_2d, r_idx):
    return [*block_2d[r_idx]]

def get_diff(list_1, list_2):
    if len(list_1) != len(list_2):
        raise Exception('error')
    diff_list = list(filter(lambda i: list_1[i] != list_2[i], range(len(list_1))))
    return len(diff_list)

def get_block_result_col(block):
    block_2d = to_add_list(block)
    col_len = len(block_2d[0])
    for i in range(col_len - 1):
        smudge = 0
        sum_ref = 2 * i + 1
        for j in range(i + 1):
            j_ref = sum_ref - j
            if not is_valid(col_len, j_ref):
                continue
            diff = get_diff(get_col_list(block_2d, j), get_col_list(block_2d, j_ref))
            smudge += diff
            if smudge > 1: 
                break
        if smudge == 1:
            return i + 1
    return 0

def get_block_result_row(block):
    block_2d = to_add_list(block)
    row_len = len(block_2d)
    # search ref line
    for i in range(row_len - 1):
        smudge = 0
        # the index sum of 2 reflected rows
        sum_ref = 2 * i + 1
        # verify a potential line
        for j in range(i+1):
            j_ref = sum_ref - j
            if not is_valid(row_len, j_ref):
                continue
            diff = get_diff(get_row_list(block_2d, j), get_row_list(block_2d, j_ref))
            smudge += diff
            if smudge > 1: 
                break
        if smudge == 1:
            # 1 based index
            return 100 * (i + 1)
    return 0
            
def get_block_result(block):
    row_res = get_block_result_row(block)
    return row_res if row_res != 0 else get_block_result_col(block)

blocks = input_d.strip().split('\n\n')
print(sum([get_block_result(b) for b in blocks]))
```
