```python
# input_d = '''
# .|...L....
# |.-.L.....
# .....|-...
# ........|.
# ..........
# .........L
# ..../.LL..
# .-.-/..|..
# .|....-|.L
# ..//.|....
# '''

# .   => empty
# | - => splitter
# \ / => mirror

dir_to_move = { 'R': (0, 1), 'L': (0, -1), 'U': (-1, 0), 'D': (1, 0)}
dir_to_dir = {   'R': {'/': 'U', 'L': 'D', '-': 'R', '|': ('U', 'D')},
                 'L': {'/': 'D', 'L': 'U', '-': 'L', '|': ('U', 'D')},
                 'U': {'/': 'R', 'L': 'L', '-': ('L', 'R'), '|': 'U'},
                 'D': {'/': 'L', 'L': 'R', '-': ('L', 'R'), '|': 'D'}}

def to_input_2d(input_d):
    rows = input_d.strip().split('\n')
    return [[*r] for r in rows]
    
def print_2d(list_2d):
    for r in list_2d:
        print(''.join(r))

def get_2d_list_len(input_2d):
    return [len(input_2d), len(input_2d[0])]

def init_2d_list(r, c, val):
    return [ [val] * c for _ in range(r)]

def get_2d_val(input_2d, pos):
    return input_2d[pos[0]][pos[1]]

def is_pos_valid(input_size, r, c):
    return r >= 0 and r < input_size[0] and c >= 0 and c < input_size[1]

def is_visited(dir, visited, pos):
    return dir in visited[pos[0]][pos[1]]

def move_beam(input_2d, input_2d_len, visited, beams, dir_to_move, dir_to_dir):
    (current_pos, current_dir) = beams.pop(0)
    if is_pos_valid(input_2d_len, current_pos[0], current_pos[1]):
        visited[current_pos[0]][current_pos[1]] += current_dir
    
    move = dir_to_move[current_dir]
    next_pos = (current_pos[0] + move[0], current_pos[1] + move[1])
    if not is_pos_valid(input_2d_len, next_pos[0], next_pos[1]):
        return
    
    next_val = get_2d_val(input_2d, next_pos)
    # todo next dir
    if next_val == '.':
        next_dir = current_dir
        if is_visited(next_dir, visited, next_pos):
            return
        beams.append((next_pos, next_dir))
    elif next_val in '/L':
        next_dir = dir_to_dir[current_dir][next_val]
        if is_visited(next_dir, visited, next_pos):
            return
        beams.append((next_pos, next_dir))
    elif next_val in '|-':
        next_dir = dir_to_dir[current_dir][next_val]
        # pass through
        if isinstance(next_dir, str):     
            if is_visited(next_dir, visited, next_pos):
                return
            beams.append((next_pos, next_dir))
        # split
        else:
            for dir in next_dir:  
                if is_visited(dir, visited, next_pos):
                    continue
                beams.append((next_pos, dir))
    else:
        raise Exception('error', next_val)

def beam_start_from(input_2d_len, current_pos, current_dir):
    # e.g., 'RD'
    visited = init_2d_list(input_2d_len[0], input_2d_len[1], '')
    
    beams = [] 
    beams.append((current_pos, current_dir))
    
    while_loop = 0
    while len(beams) > 0:
        while_loop += 1
        if while_loop > 99999:
            # for r in visited:
                # print(r)
            raise Exception("while loop", while_loop)
        
        move_beam(input_2d, input_2d_len, visited, beams, dir_to_move, dir_to_dir)
    
    count = 0
    for r in visited:
        # print(r)
        for c in r:
            if len(c) > 0:
                count += 1
    return count

input_2d = to_input_2d(input_d)
# print_2d(input_2d)
input_2d_len = get_2d_list_len(input_2d)
# print(input_2d_len)
  
# 46
# 7074
# print('challenge 1:', beam_start_from(input_2d_len, (0, -1), 'R'))


print('challenge 2:', max(
    max([beam_start_from(input_2d_len, (row_i, -1), 'R') for row_i in range(input_2d_len[0])]),
    max([beam_start_from(input_2d_len, (row_i, input_2d_len[1]), 'L') for row_i in range(input_2d_len[0])]),
    max([beam_start_from(input_2d_len, (-1, col_i), 'D') for col_i in range(input_2d_len[1])]),
    max([beam_start_from(input_2d_len, (input_2d_len[0], col_i), 'U') for col_i in range(input_2d_len[1])])))

```
