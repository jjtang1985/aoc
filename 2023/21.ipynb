```python
input_d = '''
...........
.....###.#.
.###.##..#.
..#.#...#..
....#.#....
.##..S####.
.##..#...#.
.......##..
.##.#.####.
.##..##.##.
...........
'''
import cProfile
possible_move = ((0, 1), (0, -1), (-1, 0), (1, 0))

def remove_unnecessary_visited_for_a_map(visited, map_pos, is_cound_odd):  
    ret = 0
    for (pos, is_odd_step) in list(visited.items()):
        if pos[2] == map_pos[0] and pos[3] == map_pos[1]:
            del visited[pos]
            ret += 1 if ((is_odd_step and is_cound_odd) or (not is_odd_step and not is_cound_odd)) else 0
    return ret

# return counted, due to the optimization
def update_visited(all_plot_len, visited, map_visited_count, map_completed, step, current_pos, is_cound_odd):
    ret = 0
    # visited
    if step%2 == 1:
        visited.update({current_pos: True})
    else:
        visited.update({current_pos: False})
    # map visited count
    map_pos = (current_pos[2], current_pos[3])
    multi_map_add(map_visited_count, map_pos)
    # optimization
    if all_plot_len == map_visited_count[map_pos]:
        # complete a map
        map_completed.add(map_pos)
        # print('map_completed: ++', len(map_completed))
        # no need for map_visted_count
        del map_visited_count[map_pos]
        # no need for visted
        ret += remove_unnecessary_visited_for_a_map(visited, map_pos, is_cound_odd)
    return ret

def multi_map_add(multi_map, key):
    if key in multi_map:
        multi_map.update({key: multi_map[key] + 1})
    else:
        multi_map.update({key: 1})
    return

def to_input_2d(input_d):
    rows = input_d.strip().split('\n')
    return [[*r] for r in rows]

def print_2d(list_2d):
    for r in list_2d:
        print(''.join(r))

def get_2d_list_len(input_2d):
    return [len(input_2d), len(input_2d[0])]

# def is_pos_valid(input_size, r, c):
#     return r >= 0 and r < input_size[0] and c >= 0 and c < input_size[1]

def get_2d_pos_first(input_2d, input_size, val):
    for i in range(input_size[0]):
        for j in range(input_size[1]):
            if input_2d[i][j] == val:
                return (i, j)

def get_2d_pos_all(input_2d, input_size, val):
    all_pos = [[(i, j) if input_2d[i][j] == val else None for j in range(input_size[1])] for i in range(input_size[0])]
    return list(filter(lambda e: e is not None, sum(all_pos, [])))

def is_visited(next_pos, visited):
    return next_pos in visited

def is_in_map_completed(next_map_x, next_map_y, map_completed):
    return (next_map_x, next_map_y) in map_completed

def get_next_pos(input_size, visited, map_completed, current_pos, move):
    next_pos_x = current_pos[0] + move[0]
    next_pos_y = current_pos[1] + move[1]
    next_map_x = current_pos[2]
    next_map_y = current_pos[3]

    # fix array out bound
    if next_pos_x < 0:
        next_pos_x += input_size[0]
        next_map_x -= 1
    if next_pos_x >= input_size[0]:
        next_pos_x -= input_size[0]
        next_map_x += 1
    if next_pos_y < 0:
        next_pos_y += input_size[1]
        next_map_y -= 1
    if next_pos_y >= input_size[1]:
        next_pos_y -= input_size[1]
        next_map_y += 1
    next_pos = (next_pos_x, next_pos_y, next_map_x, next_map_y)
    # visited
    if is_visited(next_pos, visited):
        return None
    # visited in completed map
    elif is_in_map_completed(next_map_x, next_map_y, map_completed):
        return None
    else:
        return next_pos
        

def get_next_pos_list(input_size, all_rock, all_plot_len, visited, map_visited_count, map_completed, step, current_pos, is_cound_odd):
    counted_from_complete_map = update_visited(all_plot_len, visited, map_visited_count, map_completed, step, current_pos, is_cound_odd)
    
    ret = []
    for move in possible_move:
        next_pos = get_next_pos(input_size, visited, map_completed, current_pos, move)
        if next_pos == None:
            continue
        if (next_pos[0], next_pos[1]) in all_rock:
            continue
        ret.append(next_pos)
    return (ret, counted_from_complete_map)

def main():

    input_2d = to_input_2d(input_d)
    input_2d_len = get_2d_list_len(input_2d)
    
    # print_2d(input_2d)
    # print(input_2d_len)
    start_pos = get_2d_pos_first(input_2d, input_2d_len, 'S')
    # print(start_pos)
    all_rock = get_2d_pos_all(input_2d, input_2d_len, '#')
    # print('# all rock', len(all_rock))
    all_plot = get_2d_pos_all(input_2d, input_2d_len, '.')
    all_plot.append(start_pos)
    all_plot_len = len(all_plot)
    # print('# all plot', all_plot_len)
    
    # {pos_x, pos_y, map_x, mapy}
    # pos_x: same as 2d
    # map_x: original map (0,0), right: (0, 1), down: (1, 0)
    current_pos_list = set()
    current_pos_list.add((start_pos[0], start_pos[1], 0, 0))
    
    # pos to is_odd_step
    # {(x, y, m_x, m_y): True}
    visited = {}
    # map to num of visited
    # {(m_x, m_y): 9}
    map_visited_count = {}
    map_completed = set()
    step = 1
    run_steps = 300
    counted_from_complete_map = 0
    
    while step <= 1 + run_steps:
        step += 1
        next_pos_list = set()
    
        # num_pos = 0
        
        for pos in current_pos_list:
            # num_pos +=1
            # if num_pos > 1000:
            #     break
            next_pos_list_and_counted = get_next_pos_list(input_2d_len, all_rock, all_plot_len, visited, map_visited_count, map_completed, step, pos, run_steps%2 == 1)
            next_pos_list.update(next_pos_list_and_counted[0])
            counted_from_complete_map += next_pos_list_and_counted[1]
        current_pos_list = next_pos_list
        # if step%50 == 0:
        #     print(len(current_pos_list))
        #     print(len(visited))
        #     print()
        
    # print(len(current_pos_list))
    # print(len(visited))
    # print(visited)
    
    #todo, not generic
    visited_is_odd = [kv[1] for kv in list(visited.items())]
    print('still in the visit', len(list(filter(lambda e: not e, visited_is_odd))))
    print('map_completed', len(map_completed))
    print('counted_from_complete_map', counted_from_complete_map)

# cProfile.run("main()")

# intial version was better than the optimized, due to the function "remove_unnecessary_visited_for_a_map"
# 6 -> 16
# 10 -> 50
# 50 -> 1594
# 100 6536
# 500 167004
# 1000 668697 9s
# 1200 ?? 13s
# 5000 ??
# 26501365
```
