```python
#36 pair, 374
input_d = '''
...#......
.......#..
#.........
..........
......#...
.#........
.........#
..........
.......#..
#...#.....
'''
directions = [[-1, 0], [1, 0], [0, -1], [0, 1]]

def copy_2d_list(list):
    ret = []
    for r in list:
        ret.append(r.copy())
    return ret

def is_pos_valid(input_size, r, c):
    return r >= 0 and r < input_size[0] and c >= 0 and c < input_size[1]

def get_2d_val(input_2d, pos):
    return input_2d[pos[0]][pos[1]]
def set_2d_val(input_2d, pos, val):
    input_2d[pos[0]][pos[1]] = val

# update neighbour pos, return #destination left
def update_neighbour(input_2d, input_size, task, tasks, step, left_dest):
    for direction in directions:
        next_pos = (task[0] + direction[0], task[1] + direction[1])
        # check array range
        if is_pos_valid(input_size, next_pos[0], next_pos[1]):
            next_val = get_2d_val(input_2d, next_pos)
            # print('next_val', next_val, next_pos)
            # unknown position updates
            if isinstance(next_val, str) and next_val in ".#":
                set_2d_val(input_2d, next_pos, step)
                tasks.append(next_pos)
            # known position, shortest path updates => never happens?
            if isinstance(next_val, int) and step < next_val:
                set_2d_val(input_2d, next_pos, step)
                print('error?')
            # found a galaxy
            if next_val == '#':
                left_dest -= 1
    return left_dest

def print_2d(list_2d):
    for r in list_2d:
        print(''.join(str(r)))
        
rows = input_d.strip().split('\n')
input_2d = [[*r] for r in rows]
input_size = [len(input_2d), len(input_2d[0])]
# print_2d(input_2d)
# print('input_size', input_size)

galaxies = []
for r in range(input_size[0]):
    for c in range(input_size[1]):
        if input_2d[r][c] == '#':
            galaxies.append((r, c))
# print('galaxies', galaxies)
# print('#galaxies', len(galaxies))

empty_galaxies = ([r for r in range(input_size[0])], [c for c in range(input_size[1])])
for galaxy in galaxies:
    if galaxy[0] in empty_galaxies[0]:
        empty_galaxies[0].remove(galaxy[0])
    if galaxy[1] in empty_galaxies[1]:
        empty_galaxies[1].remove(galaxy[1])
# print('empty_galaxies', empty_galaxies)

# updates empty galaxies starts
empty_galaxies[0].reverse()
for e_g_r in empty_galaxies[0]:
    new_line = ['.' for i in range(input_size[1])]
    input_2d.insert(e_g_r, new_line)
input_size[0] = len(input_2d)

empty_galaxies[1].reverse()
for e_g_c in empty_galaxies[1]:
    for r in range(input_size[0]):
        # print(r, e_g_c)
        input_2d[r].insert(e_g_c, '.')
input_size[1] = len(input_2d[0])
# updates empty galaxies ends

# update galaxies again
galaxies = []
for r in range(input_size[0]):
    for c in range(input_size[1]):
        if input_2d[r][c] == '#':
            galaxies.append((r, c))

# print_2d(input_2d)
print('input_size', input_size)
print('galaxies', galaxies)
print('#galaxies', len(galaxies))

# no need for the last one
res = 0
dest = galaxies.copy()
for start_g_i in range(len(galaxies) - 1):
    #todo
    # if start_g_i > 1:
    #     break
        
    input_2d_copy = copy_2d_list(input_2d)
    for started_g_i in range(start_g_i):
        set_2d_val(input_2d_copy, galaxies[started_g_i], '.')

    # print('removing:', start_g_i)
    # del dest[start_g_i]
    dest.pop(0)
    # print('start:', start_g_i, ' #dest: ' , len(dest), dest)
    left_dest = len(galaxies) - (start_g_i + 1)
    
    current_step = 0
    current_pos = galaxies[start_g_i]
    tasks = [current_pos]
    # print('init tasks', tasks)
    set_2d_val(input_2d_copy, current_pos, 0)
    
    while_loop = 0
    while True:
        while_loop += 1
        # print('current_step', current_step)
        if while_loop > 300:
            print('error 1:', while_loop)
            print_2d(input_2d_copy)
            break
        next_round_tasks = []
        for task in tasks:
            left_dest = update_neighbour(input_2d_copy, input_size, task, next_round_tasks, current_step + 1, left_dest)
            # print('left_dest', left_dest)
            # print('tasks', tasks)
            if left_dest == 0:
                break                
        if left_dest == 0:
            break
        tasks = next_round_tasks
        current_step += 1
    # print_2d(input_2d_copy)
    # print('dest', dest)
    res += sum(list(map(lambda pos: get_2d_val(input_2d_copy, pos), dest)))
print(res)
# 9723824
```
